```python
import pandas as pd
import numpy as np

print("Pandas version:", pd.__version__)
print("NumPy version:", np.__version__)
```

```python
#To create Mock Sales Data 
np.random.seed(42)  # For reproducibility

products = ['Laptop', 'Mouse', 'Keyboard', 'Monitor', 'Headphones',
            'Webcam', 'USB Hub', 'SSD', 'RAM', 'Charger']

regions = ['North', 'South', 'East', 'West']

dates = pd.date_range(start='2024-01-01', end='2024-06-30', freq='D')

#random data of 100 rows 
n = 100
data = {
    'product':  np.random.choice(products, n),
    'region':   np.random.choice(regions, n),
    'date':     np.random.choice(dates, n),
    'quantity': np.random.randint(1, 50, n),
    'price':    np.round(np.random.uniform(10, 500, n), 2)
}

sales_df = pd.DataFrame(data)
sales_df['date'] = pd.to_datetime(sales_df['date']).dt.date  # sirf date rakhna

# CSV save karo
sales_df.to_csv('sales_data.csv', index=False)

print("sales_data.csv successfully created!")
print("Shape:", sales_df.shape)
```

```python
# CSV load
df = pd.read_csv('sales_data.csv')

print("=" * 45)
print("FILE LOADED SUCCESSFULLY")
print("=" * 45)

print("\nFirst 5 Rows:")
print(df.head())

print("\nShape (rows, columns):", df.shape)

print("\nColumn Names:", df.columns.tolist())

print("\nData Types:")
print(df.dtypes)

print("\nMissing Values:")
print(df.isnull().sum())

print("\nBasic Statistics:")
print(df.describe())
```

```python
# Only filter 'North' region data
region_filter = 'North'
df_north = df[df['region'] == region_filter]

print(f"Filter Applied: Region = '{region_filter}'")
print(f"Original Data Rows: {len(df)}")
print(f"Filtered Data Rows: {len(df_north)}")

print(f"\nFiltered Data (First 5 Rows):")
print(df_north.head())
```

```python
# Revenue = quantity * price
df_north = df_north.copy()  #To avoid warning 
df_north['revenue'] = df_north['quantity'] * df_north['price']
df_north['revenue'] = df_north['revenue'].round(2)

print("Revenue Column Added!")
print("\nUpdated DataFrame (First 5 Rows):")
print(df_north[['product', 'region', 'quantity', 'price', 'revenue']].head())

print("\nNew Shape:", df_north.shape)
print("Total Revenue (North):", round(df_north['revenue'].sum(), 2))
```

```python
# group revenue according to total products
product_revenue = df_north.groupby('product')['revenue'].sum().reset_index()
product_revenue.columns = ['product', 'total_revenue']
product_revenue['total_revenue'] = product_revenue['total_revenue'].round(2)

# Sorting Descending order 
product_revenue = product_revenue.sort_values('total_revenue', ascending=False)

print("Total Revenue by Product (North Region):")
print(product_revenue)
print("\nShape:", product_revenue.shape)
```

```python
# Top 5 products by revenue
top5 = product_revenue.head(5)

print("Top 5 Products by Revenue (North Region):")
print("=" * 40)
print(top5.to_string(index=False))
print("=" * 40)
print(f"\nTop Product: {top5.iloc[0]['product']} with Revenue: {top5.iloc[0]['total_revenue']}")
```

```python
# Product metadata table — extra info like category and brand
metadata = pd.DataFrame({
    'product':  ['Laptop', 'Mouse', 'Keyboard', 'Monitor', 'Headphones',
                 'Webcam', 'USB Hub', 'SSD', 'RAM', 'Charger'],
    'category': ['Computer', 'Accessory', 'Accessory', 'Display', 'Audio',
                 'Accessory', 'Accessory', 'Storage', 'Hardware', 'Accessory'],
    'brand':    ['Dell', 'Logitech', 'Logitech', 'Samsung', 'Sony',
                 'Logitech', 'Anker', 'Samsung', 'Kingston', 'Anker']
})

print("Product Metadata Table:")
print(metadata)
print("\nShape:", metadata.shape)
```

```python
#  merging of Top 5 products to metadata
final_df = pd.merge(top5, metadata, on='product', how='left')

print("Merged Table (Top 5 + Metadata):")
print("=" * 55)
print(final_df.to_string(index=False))
print("=" * 55)
print("\nFinal Shape:", final_df.shape)
print("Columns:", final_df.columns.tolist())
```

```python
#  Export Final result
output_filename = 'top5_products_north.csv'
final_df.to_csv(output_filename, index=False)

print(f"File exported successfully: '{output_filename}'")
print(f"\nFinal Exported Data:")
print(final_df)
print(f"\nTotal Rows Exported: {len(final_df)}")
print(f"Total Columns: {len(final_df.columns)}")
```
